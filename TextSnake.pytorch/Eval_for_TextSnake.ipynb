```python
import numpy as np
import pandas
import scipy.misc
import random
import sys
if '/opt/ros/kinetic/lib/python2.7/dist-packages' in sys.path:
    sys.path.remove('/opt/ros/kinetic/lib/python2.7/dist-packages')
import cv2
from matplotlib import pyplot as plt
import time
import os
```

```python
file_folder = "./save"
train_name = "example"
val_loss = ['loss_result.txt', 'tr_loss_result.txt', 'tcl_loss_result.txt',\
            'sin_result.txt', 'cos_result.txt', 'radii_result.txt']

file_path = os.path.join(file_folder, train_name)
```

```python

for i in range(len(val_loss)):
    _list = []
    count = 0
    with open(os.path.join(file_path, val_loss[i]),'r') as fp:
        all_lines = fp.readlines()
        for _str in all_lines:
            if count % 10 == 0:
                _str = _str.split("_")
                _list.append(float(_str[-1][:-1]))
            count += 1
    x = np.arange(0,len(_list),1)       
    plt.figure()
    plt.title(val_loss[i][:-4])
    plt.plot(x, _list, color = 'red', label = val_loss[i][:-11])
    plt.legend()
    plt.xlim(0,len(_list))
    plt.ylim(0,0.6)
```
